```python
import pandas as pd
```

```python
# Selecting specific columns due to memory constraints
file_path = 'Resources/Covid-19/Data/COVID-19_Case_Surveillance_Public_Use_Data_with_Geography.csv'

cols_list = [
    'res_county',
     'age_group', 
     'sex', 
     'current_status', 
     'symptom_status',
     'hosp_yn',
     'icu_yn',
     'death_yn',
     'underlying_conditions_yn'
]

mlm = pd.read_csv(file_path, usecols=cols_list)
```

```python
mlm
```

```python
# Dropping NA's, already have so much data
mlm = mlm.dropna()
mlm
```

```python
# Only keeping rows where death_yn where death is 'Yes' or 'No'
mlm = mlm.loc[((mlm['death_yn'] == 'Yes') | (mlm['death_yn'] == 'No')) & 
              
              # # Only keeping rows where underlying conditions is 'Yes' or 'No'
              ((mlm['underlying_conditions_yn'] == 'Yes') | (mlm['underlying_conditions_yn'] == 'No'))]
mlm
```

```python
mlm.symptom_status.unique()
```

```python
# Only keeping rows with 'Symptomatic' or 'Asymptomatic'
mlm = mlm.loc[(mlm['symptom_status'] == 'Symptomatic') | (mlm['symptom_status'] == 'Asymptomatic')]
mlm
```

```python
mlm.icu_yn.unique()
```

```python
# Keeping icu-yn 'yes' or 'no'
mlm = mlm.loc[(mlm['icu_yn'] == 'Yes') | (mlm['icu_yn'] == 'No')]
```

```python
# keep only confirmed cases
mlm = mlm.loc[mlm['current_status'] == 'Laboratory-confirmed case']
mlm
```

```python
# We can drop the current_status column
mlm.drop(columns='current_status', inplace=True)
```

```python
# export df to csv file
file_path = 'Resources/Covid-19/Cleaned_Data/machine_learning_data.csv'
mlm.to_csv(file_path)
```

```python
# Import and clean census data
file_path = 'Resources/Covid-19/Data/pop_by_county_us_census_2020.csv'
Census_county_pop = pd.read_csv(file_path)
```

```python
# Getting County and population
Census_county_pop = Census_county_pop[['NAME','P1_001N']]
Census_county_pop
```

```python
# Renaming columns
Census_county_pop = Census_county_pop.rename(columns={"NAME":"County_Name", "P1_001N": "County_pop"})
```

```python
# Dropping first row
Census_county_pop = Census_county_pop.iloc[1: , :]
Census_county_pop
```

```python
split_df = Census_county_pop['County_Name'].str.split(',',expand=True)
```

```python
split_df
```

```python
Census_county_pop = split_df.join(Census_county_pop)
Census_county_pop
```

```python
Census_county_pop.drop(columns= 'County_Name', inplace=True)
```

```python
# Renaming Columns county and state
Census_county_pop.rename(columns={0:'County_name',1:'State'},inplace=True)
```

```python
# Getting rid of county
Census_county_pop['County_name'] = Census_county_pop['County_name'].str.replace(' County', '')
Census_county_pop
```

```python
# Get county names to match so they can be merged with mlm dataframe
Census_county_pop['County_name'] = Census_county_pop['County_name'].str.upper()
Census_county_pop
```

```python
# export to csv
file_path = 'Resources/Covid-19/Cleaned_Data/County_clean.csv'
Census_county_pop.to_csv(file_path)
```

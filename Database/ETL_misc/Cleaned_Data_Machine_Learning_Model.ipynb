```python
import pandas as pd
```

```python
# Selecting specific columns due to memory constraints
file_path = '../Resources/Covid-19/Data/COVID-19_Case_Surveillance_Public_Use_Data_with_Geography.csv'

cols_list = [
    'res_county',
    'res_state',
     'age_group', 
     'sex', 
     'current_status', 
     'symptom_status',
     'hosp_yn',
     'icu_yn',
     'death_yn',
     'underlying_conditions_yn'
]

mlm = pd.read_csv(file_path, usecols=cols_list)
```

```python
# Dropping NA's, already have so much data
mlm = mlm.dropna()
mlm
```

```python
# Only keeping rows where death_yn where death is 'Yes' or 'No'
mlm = mlm.loc[((mlm['death_yn'] == 'Yes') | (mlm['death_yn'] == 'No')) & 
              
              # # Only keeping rows where underlying conditions is 'Yes' or 'No'
              ((mlm['underlying_conditions_yn'] == 'Yes') | (mlm['underlying_conditions_yn'] == 'No')) &
             
              # Only keeping rows with 'Symptomatic' or 'Asymptomatic'
              ((mlm['symptom_status'] == 'Symptomatic') | (mlm['symptom_status'] == 'Asymptomatic')) &
             
              # Keeping icu-yn 'yes' or 'no'
              ((mlm['icu_yn'] == 'Yes') | (mlm['icu_yn'] == 'No')) &
             
              # keep only confirmed cases
              ((mlm['current_status'] == 'Laboratory-confirmed case')) &
              
              # keep only Male or Female
              ((mlm['sex'] == 'Male') | (mlm['sex'] == 'Female')) &
             
              # keep only hosp_yn
              ((mlm['hosp_yn'] == 'Yes') | (mlm['hosp_yn'] == 'No'))
]

mlm
```

```python
# We can drop the current_status column
mlm.drop(columns='current_status', inplace=True)
```

```python
# Stripping res_county
mlm['res_county'] = mlm['res_county'].str.strip()
```

```python
# stripping res_state
mlm['res_state'] = mlm['res_state'].str.strip()
```

```python
# Creating join column
mlm['joins'] = mlm['res_county'] + ', ' + mlm['res_state']
mlm
```

```python
# export df to csv file
file_path = 'Resources/Covid-19/Cleaned_Data/machine_learning_data.csv'
mlm.to_csv(file_path)
```

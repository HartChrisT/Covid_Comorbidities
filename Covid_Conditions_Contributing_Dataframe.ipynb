```python
# Imports
import pandas as pd
import numpy as np
import csv
```

```python
# Load in CSV files
file_path = 'Resources/Covid-19/Data/Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age__Provisional_2020-2022.csv'
Covid_contributing_conditions = pd.read_csv(file_path)
Covid_contributing_conditions
```

```python
Covid_contributing_conditions.columns.tolist()
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions.loc[Covid_contributing_conditions['Group'] == 'By Month']
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions_monthly.drop(columns=['Flag', 'Data As Of', 'End Date', 'Group', 'Year', 'Month', 'ICD10_codes'])
```

```python
Covid_contributing_conditions_monthly
```

```python
Covid_contributing_conditions_monthly['Start Date'] = pd.to_datetime(Covid_contributing_conditions_monthly['Start Date'])
```

```python
Covid_contributing_conditions_monthly= Covid_contributing_conditions_monthly.drop(columns='Number of Mentions')
```

```python
Covid_contributing_conditions_monthly
```

```python
Covid_contributing_conditions_monthly['Start Date'] = Covid_contributing_conditions_monthly['Start Date'].dt.strftime('%m/%Y')
```

```python
Covid_contributing_conditions_monthly
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions_monthly.dropna()
```

```python
Covid_contributing_conditions_monthly
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions_monthly.reset_index()
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions_monthly.drop(columns=['index'])
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions_monthly.rename(columns= {'Start Date':'Month',
                                                                                              'Condition Group': 'Condition_Group',
                                                                                              'Age Group': 'Age_Group',
                                                                                              'COVID-19 Deaths': 'COVID-19_Deaths'})
```

```python
Covid_contributing_conditions_monthly
```

```python
Covid_contributing_conditions_monthly = Covid_contributing_conditions_monthly.astype({'COVID-19_Deaths': 'int32'}).dtypes
```

```python
file_path = 'Resources/Covid-19/Cleaned_Data/Covid_contributing_conditions_monthly.csv'
Covid_contributing_conditions_monthly.to_csv(file_path)
```
